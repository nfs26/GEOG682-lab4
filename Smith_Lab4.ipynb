```python
import rasterio
from rasterio import plot
import matplotlib.pyplot as plt
```

```python
# load image1 and image2
sent1 = rasterio.open("image1.tif")
sent2 = rasterio.open("image2.tif")
```

```python
plot.show(sent1, cmap='terrain')
```

```python
plot.show(sent2, cmap='terrain')
```

```python
sent1.crs
#CRS = 32618
```

```python
#plot each band of the image individually
fig, (axr, axg, axb, axn) = plt.subplots(1,4, figsize=(20,5))
plot.show((sent1, 3), ax=axr, cmap='Reds', title='red channel')
plot.show((sent1, 2), ax=axg, cmap='Greens', title='green channel')
plot.show((sent1, 1), ax=axb, cmap='Blues', title='blue channel')
```
